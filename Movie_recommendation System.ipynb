```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj): #this convert in String to the actual dataset
        l.append(i['name'])
    return l
```

```python
movies_df['genres']=movies_df['genres'].apply(convert)
movies_df['keywords']=movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
movies_df.head()
```

```python
def convert3(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            l.append(i['name'])
        else:
            break
        return l
```

```python
movies_df['crew']=movies_df['crew'].apply(convert3)
```

```python
movies_df.head()
```

```python
movies_df['overview'][0]
```

```python
movies_df['overview']=movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df
```

```python
movies_df['genres']=movies_df['genres'].apply(lambda x:[i.replace(" "," ") for i in x])
movies_df['keywords']=movies_df['keywords'].apply(lambda x:[i.replace(" "," ") for i in x])
movies_df['cast']=movies_df['cast'].apply(lambda x:[i.replace(" "," ") for i in x])
#movies_df['crew']=movies_df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_df.head()
```

```python
movies_df['tags']=movies_df['overview']+movies_df['genres']+movies_df['keywords']+movies_df['cast']+movies_df['crew']
```

```python
movies_df.head()
```

```python
# creating a new data frame
new_df=movies_df[['movie_id','title','tags']]
new_df
```

```python
# for removing the array brackets
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join(map(str, x)) if isinstance(x, list) else str(x))

# the tags cantains some float elements 
```

```python
new_df
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(100)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors)
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommand(movies):
    index=new_df[new_df['title']==movies].index[0]
    distance =sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_df.iloc[i[0]].title)
```
